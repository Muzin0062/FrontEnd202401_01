## 命令と入力データを分離して制御する

```python
# 入力データを分離してテンプレート化する
def generate_recipe_gemini( dish:str ) -> str:
  # 入力データを引数で受け取りLLMに渡すプロンプトを作成して実行する関数
  prompt = f"以下の料理のレシピを考えてください\n{dish}"
  print( prompt ) # 組み立てたプロンプトを画面に表示

  # モデルを生成してプロンプトを引き渡し、応答を受け取る
  model = genai.GenerativeModel("gemini-flash-latest")
  response = model.generate_content( prompt )
  return response.text
# 関数定義終わり
```

## 完成した関数を呼び出して利用する

```python
# コロッケの作り方
recipe_gemini = generate_recipe_gemini("コロッケ")
print( recipe_gemini )
```

## ユーザーが指定した文脈をプロンプトにはめこんでいく

```python
def generate_spicy_recipe_gemini( dish:str, amount:str, fav:str) -> str :
  context = f"前提条件:\n分量:{amount}\n味の好み:{fav}"
  prompt = f"前提条件を踏まえて、以下のレシピを考えてください。\n{context}\n料理名:{dish}"
  # プロンプトを画面に表示
  print( prompt )
  model = genai.GenerativeModel("gemini-flash-latest")
  response = model.generate_content( prompt )
  return response.text
# 関数終わり
```

### 関数を呼び出す

```python
recipe_gemini = generate_spicy_recipe_gemini("麻婆豆腐","2人分","辛口")
print( recipe_gemini )
```

## JSONフォーマットの出力形式を指定する

```python
import json
from IPython.display import JSON # JSON形式を画面で見やすく表示させる
```

### 再利用性を高めるために関数で定義

```python
def generate_json_recipe_gemini( dish:str ) -> dict:
  """ 料理名を引数でもらい、その料理のレシピのLLMの応答をJSONで返す """

  prompt = f""" \
  ユーザーが入力した料理のレシピを考えてください。
  出力は以下のJSON形式にしてください。
  {
      {
          "材料":["材料1","材料2"],
          "手順":["手順1","手順2"]
      }
  }
  料理名:{dish}
   """
  model = genai.GenerativeModel("gemini-flash-latest")
  response = model.generate_content( prompt,generation_config={"response_mime_type":"application/json"} )
  # json文字列をオブジェクトにしてリターン
  return json.loads(response.text)
  # 関数終わり
```

### 関数を呼び出し実行して、JSONオブジェクトの応答を表示

```python
json_recipe_gemini = generate_json_recipe_gemini( "欧風ハンバーグ" )
display( JSON(json_recipe_gemini).data )
```

## FewShotプロンプティング

```python
prompt = """
主な材料から、幼児向けの分かりやすい楽しそうな料理名を創作してください。
材料:鶏肉、玉ねぎ、卵、ごはん
料理名:ふわふわとろ～りおやこどん
材料:豚肉、キムチ、豆腐
料理名:うまから～いびっくりキムチチゲ
材料:トマト、ナス、ズッキーニ、パプリカ
料理名:
"""

response = model.generate_content( prompt )
print( "--- プロンプト ---" )
print( prompt )
print( "\n --- AIからの応答 ---" )
print( response.text )
```

# Zero shot Chain of Thought プロンプティング

```python
model = genai.GenerativeModel( "gemini-flash-latest" )
prompt = "名探偵コナンの招待がばれないのはなぜでしょうか。ステップバイステップで5段間にわけて小学生にも分るように説明して"
response = model.generate_content( prompt )
print( "--- プロンプト ---" )
print( prompt )
print( "\n--- AIからの回答 ---" )

import textwrap
wrapped_text = textwrap.fill( response.text, width=40 )
print( wrapped_text )
```
